```python
import random

trials = 10000
count_heads_and_six = 0

for _ in range(trials):
    coin = random.choice(['Heads', 'Tails'])
    dice = random.randint(1, 6)
    if coin == 'Heads' and dice == 6:
        count_heads_and_six += 1

prob_independent = count_heads_and_six / trials
print("Experimental Probability (Heads AND 6):", prob_independent)

theoretical_independent = (1/2) * (1/6)
print("Theoretical Probability (Heads AND 6):", theoretical_independent)

count_two_red = 0
marbles = ['Red']*5 + ['Blue']*5

for _ in range(trials):
    random.shuffle(marbles)
    pick1 = marbles.pop()
    pick2 = marbles.pop()
    if pick1 == 'Red' and pick2 == 'Red':
        count_two_red += 1
    marbles = ['Red']*5 + ['Blue']*5

prob_dependent = count_two_red / trials
print("Experimental Probability (2 Red Marbles):", prob_dependent)

theoretical_dependent = (5/10) * (4/9)
print("Theoretical Probability (2 Red Marbles):", theoretical_dependent)
```
