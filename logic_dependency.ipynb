```python
import random

trials = 10000

# Counters
count_heads_and_6 = 0
count_two_red = 0

for _ in range(trials):

    # INDEPENDENT EVENT
    coin = random.choice(["H", "T"])
    die = random.randint(1, 6)

    if coin == "H" and die == 6:
        count_heads_and_6 += 1

    # DEPENDENT EVENT
    bag = ["R"] * 5 + ["B"] * 5
    first_pick = random.choice(bag)
    bag.remove(first_pick)
    second_pick = random.choice(bag)

    if first_pick == "R" and second_pick == "R":
        count_two_red += 1

exp_independent = count_heads_and_6 / trials
exp_dependent = count_two_red / trials

print("----- INDEPENDENT EVENT -----")
print("Experimental Probability (Heads AND 6):", exp_independent)
print("Theoretical Probability:", 1/12)

print("\n----- DEPENDENT EVENT -----")
print("Experimental Probability (2 Red Marbles without replacement):", exp_dependent)
print("Theoretical Probability:", 2/9)
```
